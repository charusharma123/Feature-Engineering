1. What is a Parameter?
- A parameter in machine learning is a configuration variable that is internal to the model and whose value can be estimated from data. Parameters are crucial for model performance as they define how input data is transformed into the desired output. For example, in a linear regression model, the slope and intercept of the line are parameters that the algorithm learns from the training data.

2. Whar is correlation? What does negative correlation mean?
- Correlation is a statistical technique used to measure and describe the strength and direction of a relationship between two variables. A correlation coefficient ranges from -1 to 1, where +1 indicates a perfect positive correlation, -1 a perfect negative correlation, and 0 indicates no correlation.

Negative correlation occurs when two variables move in opposite directions. As one increases, the other decreases. For example, the number of hours spent watching TV and academic performance may have a negative correlation.

3. Define Machine Learning. What are the main components in Machine Learning?
- Machine Learning (ML) is a subset of artificial intelligence (AI) that enables systems to automatically learn and improve from experience without being explicitly programmed. Key components include:

 * Data: The raw material used for training.

 * Model: The algorithm that processes data.

 * Loss Function: Measures prediction error.

 * Optimizer: Updates parameters to minimize loss.

 * Evaluation Metrics: Assess model performance.

4. How does loss value help in determining whether the model is good or not?
- The loss value indicates how far the model's predictions are from the actual outcomes. A smaller loss signifies better performance. During training, the goal is to minimize this value. Monitoring loss helps detect overfitting or underfitting.

5. What are continuous and catagorical variables?

Categorical variables represent distinct groups or categories (e.g., gender, color).

6. How do we handle categorical variables in Machine Learning? WHat are the common techniques?
- To use categorical variables in ML models, we convert them to numerical formats:

 * Label Encoding assigns each category a unique integer.

 * One-Hot Encoding creates binary columns for each category.

 * Ordinal Encoding ranks categories in a meaningful order.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
- An optimizer is an algorithm that adjusts model parameters to minimize the loss function. Types:
 * Gradient Descent: Basic optimizer using gradient of loss function
 * SGD (Stochastic GD): Uses random samples
 * Adam: Adaptive Moment Estimation (combines momentum and RMSprop)
 * RMSprop: Adapts learning rate per parameter
 * Adagrad: Adapts learning rate based on parameter frequency
